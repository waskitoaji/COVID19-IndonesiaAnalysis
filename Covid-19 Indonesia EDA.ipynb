```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_columns', None)
```

# importing covid 19 indonesia data (source : kaggle)

```python
data=pd.read_csv('data/covid_19_indonesia_time_series_all.csv')
data.head(10)
```

```python
data['Location'].value_counts()
```

```python
data.info()
```

```python
data.drop(data.loc[data['Location']=='Indonesia'].index, inplace=True)
data['Location'].value_counts()
```

```python
data.tail()
```

```python
data['Date']= pd.to_datetime(data['Date'])
```

```python
df=data[['Date','Location','New Cases','New Deaths','New Recovered','New Active Cases','Total Cases','Total Deaths',
        'Total Recovered','Total Active Cases','Population','Longitude','Latitude','Growth Factor of New Cases',
         'Growth Factor of New Deaths']]
```

```python
df.head()
```

# Replace NaN value with average value

```python
avg_growthdeaths = df["Growth Factor of New Deaths"].astype("float").mean()
print("Average of Growth Factor of New Deaths = ", avg_growthdeaths)
avg_growthcase = df["Growth Factor of New Cases"].astype("float").mean()
print("Average of Growth Factor of New Cases = ", avg_growthcase)
df["Growth Factor of New Deaths"].replace(np.nan, avg_growthdeaths, inplace = True)
df["Growth Factor of New Cases"].replace(np.nan, avg_growthcase, inplace = True)
```

```python
df.head(10)
```

```python
df_case=df.groupby(['Date'])['Total Cases'].agg('sum')
df_case=pd.DataFrame(df_case)
df_case.head()
```

```python
date= df['Date'].unique()
```

```python
df_case['Date']=date
```

```python
df_case
```

```python
X=df_case['Date']
Y=df_case['Total Cases']
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 6))
plt.scatter(X,Y)
plt.xticks(rotation=45)
```

# Curve fitting using logistic growth model for predict end of pandemic

```python
from scipy.optimize import curve_fit
```

```python
#country_data['GrowthFactor'] = growth_factor(country_data['Confirmed'])

# we will want x_data to be the number of days since first confirmed and the y_data to be the confirmed data. This will be the data we use to fit a logistic curve
x_data = range(len(df_case['Date']))
y_data = df_case['Total Cases']

def log_curve(x, k, x_0, ymax):
    return ymax / (1 + np.exp(-k*(x-x_0)))

# Fit the curve
popt, pcov = curve_fit(log_curve, x_data, y_data)
estimated_k, estimated_x_0, ymax= popt
```

```python
# Plot the fitted curve
k = estimated_k
x_0 = estimated_x_0
y_fitted = log_curve(range(0,500), k, x_0, ymax)
print(k, x_0, ymax)
#print(y_fitted)
y_data.tail()
```

```python
# Plot everything for illustration
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(range(0,500), y_fitted, '--', label='Predicted Case')
ax.plot(x_data, y_data, 'o', label='Real Case')
ax.set_xlabel('Days')
ax.set_ylabel('Total Cases')
ax.legend()
#ax.set_ylim(0,300000)
plt.show()
print('Day 0 is the earliest date in dataset')
```

# Evaluation model

```python
import numpy as np
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_data, y_fitted[:225])
#fitness of the model we evaluate using %RMSE
PercentRMS=100*np.sqrt(mse)/y_data[-1]
print('Percent RMS : ',PercentRMS, '%')
```
